#データ理解

```
import pandas as pd

# 列名を追加
u_cols = ['user_id', 'age', 'sex', 'occupation', 'zip_code']
# データの読込
users = pd.read_csv(path + 'u.user', sep='|', names=u_cols)
# データの先頭行表示
users.head()
```

```
r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp', 'date']
# データの読込
ratings = pd.read_csv(path + 'u.data', sep='\t', names=r_cols)
# データの先頭行表示
ratings.head()
```

 良いデータとは誰がいつどんな、どのようにが理解できる物

```
m_cols = ['movie_id', 'title', 'release_date', 'video_release_date', 'imdb_url']
# データの読込
movies = pd.read_csv(path + '/u.item', sep='|', names=m_cols, usecols=range(5), encoding='latin1')
# データの先頭行表示
movies.head()
```

```
movie_rating = pd.merge(movies, ratings)
movie_rating
```

```
lens = pd.merge(movie_rating, users)
lens
```
